```
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('distiluse-base-multilingual-cased')
```

```
import pandas as pd

data = data = pd.read_csv('nsmc/ratings.txt', sep='\t')
train_data = pd.read_csv('nsmc/ratings_train.txt', sep='\t')
test_data = pd.read_csv('nsmc/ratings_test.txt', sep='\t')

train_data.head()
```

```
from sklearn.cluster import KMeans

# Corpus with example sentences
corpus = train_data['document'].values.tolist()[:100]

corpus_embeddings = model.encode(corpus, convert_to_tensor=True)


# Then, we perform k-means clustering using sklearn:
from sklearn.cluster import KMeans

num_clusters = 2
clustering_model = KMeans(n_clusters=num_clusters)
clustering_model.fit(corpus_embeddings)
cluster_assignment = clustering_model.labels_
```

```
part_df = train_data[: 100]
```

```
part_df['cluster'] = cluster_assignment
```
